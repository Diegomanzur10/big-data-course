## Utilizar WHERE para recuperar datos
Supongamos que estamos interesados en los registros correspondientes únicamente al año 1981. Utilizar un WHERE es bastante sencillo para un criterio simple como éste.

```python
%%sql 
SELECT RCH, YR, MO, FLOW_INcms, FLOW_OUTcms From rch 
WHERE YR=1981
```

## Más combinaciones de filtros
También podemos filtrar aún más los datos con los operadores de != o <> para obtener datos excepto 1981..

```python
%sql SELECT RCH, YR, MO, FLOW_INcms, FLOW_OUTcms From rch WHERE YR<>1981 and RCH=10 and MO=6
```

Podemos filtrar los datos a meses específicos utilizando la sentencia OR. Por ejemplo, queremos comprobar los datos de los meses 3, 6 y 9. Sin embargo, tenemos que usar () para hacerlos como una condición. Es un truco. Puedes intentarlo.

```python
%sql SELECT RCH, YR, MO, FLOW_INcms, FLOW_OUTcms From rch WHERE YR>2009 and RCH=10 and (MO=3 or MO=6 or MO=9 or MO=12)
```

Or we can simplify the above filter using the IN statement.

```python
%sql SELECT RCH, YR, MO, FLOW_INcms, FLOW_OUTcms From rch WHERE YR>2009 and RCH=10 and MO in (3, 6, 9, 12)
```

O los meses que **NO** están en 3, 6, 9, 12

```python
%sql SELECT RCH, YR, MO, FLOW_INcms, FLOW_OUTcms From rch WHERE YR>2009 and RCH=10 and MO NOT IN (3,6,9,12)
```

## Filtrar con operadores matemáticos
Por ejemplo, podemos utilizar el operador de módulo (%) para filtrar las MO.

```python
%sql SELECT RCH, YR, MO, FLOW_INcms, FLOW_OUTcms From rch WHERE YR>2009 and RCH=10 and MO % 3 = 0
```

## Calcular la diferencia entre dos columnas
Por ejemplo, nos interesa la diferencia entre la evapotranspiración potencial (PET_mm) y la precipitación (PREC_mm). Es el llamado Déficit de evapotranspiración potencial (PED). El PED puede considerarse como un índice de sequía. Es la diferencia entre la cantidad de agua que potencialmente podría perderse del suelo a través de la evapotranspiración y la cantidad realmente disponible. Cuando el PED es elevado, las plantas no disponen de toda la cantidad de agua que necesitan para crecer.

```python
%%sql sqlite://
SELECT YR, MO,  
PREC_mm as Precipitation, 
PET_mm as PET, 
PET_mm-PREC_mm as PED 
From watershed_monthly LIMIT 10
```

También podemos calcular la relación entre el PED y la precipitación.

```python
%%sql sqlite://
SELECT YR, MO,  PREC_mm as Precipitation, 
PET_mm as PET, 
(PET_mm-PREC_mm)/PREC_mm*100.0 as PED_Ratio 
From watershed_monthly LIMIT 10
```

## Realice algunas estadísticas con las funciones de agregación
Hasta ahora, hemos realizado operaciones matemáticas a través de las columnas de cada fila de una tabla. También podemos calcular un resultado a partir de valores dentro de la misma columna utilizando la función de agregación, que calcula un único resultado a partir de múltiples entradas. Dos de las funciones de agregación más utilizadas en el análisis de datos son avg() y sum().

### Promedio
avg - calcula la media de todos los valores de esa columna (omite los valores nulos).

```python
%%sql sqlite://
SELECT avg(FLOW_INcms), avg(FLOW_OUTcms) 
From rch
```

### Suma
sum - calcula la suma de los valores de esa columna (omite los valores nulos).

```python
%%sql sqlite://
SELECT sum(FLOW_INcms), sum(FLOW_OUTcms) 
From rch
```

### valores extremos
max - calcula el valor máximo de esa columna (omite los valores nulos).

min - calcula el valor mínimo en esa columna (omite los valores nulos).

```python
%%sql sqlite://
SELECT min(FLOW_INcms), max(FLOW_OUTcms) 
From rch
```

### Calcular por nosotros mismos
Podemos calcular algunos valores mediante la combinación de estos operadores matemáticos. Por ejemplo, podemos calcular los valores medios de Flow_In y Flow_Out.

```python
%%sql sqlite://
SELECT sum(FLOW_INcms)/COUNT(FLOW_INcms) as AVG_FlowIn, 
sum(FLOW_OUTcms)/COUNT(FLOW_OUTcms) as AVG_FlowOut
From rch
```

## Resumen
La agregación de datos (también conocida como agrupar, resumir o agrupar datos) consiste en crear algún tipo de total a partir de un número de registros. La suma, el mínimo, el máximo, el recuento y la media son operaciones de agregación habituales.

De hecho, el ejemplo anterior no presenta la verdadera potencia de estas funciones de agregación. Sólo serán más potentes cuando se utilicen con las cláusulas GROUP BY y ORDER BY.

## Agregación de datos con GROUP BY y ORDER BY
La cláusula GROUP BY es una cláusula opcional de la sentencia SELECT. La cláusula GROUP BY un grupo seleccionado de filas en filas de resumen por valores de una o más columnas.

La cláusula GROUP BY devuelve una fila por cada grupo. Para cada grupo, puede aplicar una función de agregación como MIN, MAX, SUM, COUNT o AVG para proporcionar más información sobre cada grupo.

### Agrupación de datos
Tomemos como ejemplo la tabla de rch

```python
%sql SELECT * From rch LIMIT 3
```

## Comprobar los valores únicos
En primer lugar, echemos un vistazo al número de RCH. Podemos utilizar la palabra clave DISTINCT junto con la sentencia SELECT para eliminar todos los registros duplicados y obtener sólo los registros únicos.


```python
%%sql sqlite://
SELECT COUNT(DISTINCT RCH) AS nRCH
FROM rch
```

También podemos utilizar la cláusula **GROUP BY** para obtener una salida más limpia, con menos filas - sólo se devuelven los valores únicos.

```python
%%sql sqlite://
SELECT RCH
FROM rch
GROUP BY RCH
```

## Utilizar funciones de agregación en los grupos
Podemos obtener más detalles mediante la agregación de datos en grupos que en columnas completas.

```python
%%sql sqlite://
SELECT RCH, AVG(FLOW_INcms), AVG(FLOW_OUTcms)
FROM rch
GROUP BY RCH
```

## Ordenar/clasificar registros
En primer lugar, vamos a comprobar los años y meses con el máximo FLOW_INcms

```python
%%sql sqlite://
SELECT RCH, YR, MO, MAX(FLOW_INcms)
FROM rch
GROUP BY RCH
```

Es obvio que las columnas de año y mes no están en una ordenación natural. Este es un buen momento para sacar a relucir el operador ORDER BY, que se puede poner al final de un estado SQL (después de cualquier WHERE y GROUP BY). Podemos ordenar los resultados de la consulta por año y luego por mes.

```python
%%sql sqlite://
SELECT RCH, YR, MO, ROUND(MAX(FLOW_INcms),2)
FROM rch
GROUP BY RCH
ORDER BY YR, MO
```

Por defecto, el orden se realiza con el operador ASC, que ordena los datos de forma ascendente. Podemos ordenar en orden descendente aplicando el operador DESC.

```python
%%sql sqlite://
SELECT RCH, YR, MO, ROUND(MAX(FLOW_INcms),2)
FROM rch
GROUP BY RCH
ORDER BY YR DESC, MO
```

## Filtrar datos sobre grupos con la cláusula HAVING
A veces, podemos querer filtrar los registros basándonos en un grupo o en un valor agregado. Aunque el primer instinto podría ser utilizar una sentencia WHERE, en realidad esto no funcionará porque el WHERE filtra los registros y no filtra las agregaciones. Por ejemplo, intentamos utilizar un WHERE para filtrar los resultados en los que MAX(FLOW_INcms) es mayor que 3000. Esto obtendrá un sqlite3.OperationalError de uso incorrecto del agregado.

```python
%%sql sqlite://
SELECT RCH, YR, MO, MAX(FLOW_INcms) as MAX_FLOWIN
FROM rch
WHERE MAX_FLOWIN > 3000.0
GROUP BY RCH
ORDER BY YR DESC, MO
```

En este caso, podemos utilizar la cláusula HAVING para especificar una condición de filtro para un grupo o un agregado. La cláusula HAVING es una cláusula opcional de la sentencia SELECT. A menudo se utiliza la cláusula HAVING con la cláusula GROUP BY. La cláusula GROUP BY agrupa un conjunto de filas en un conjunto de filas de resumen o grupos. A continuación, la cláusula HAVING filtra los grupos en función de las condiciones especificadas.

Hay que tener en cuenta que la cláusula HAVING debe seguir estrictamente la cláusula GROUP BY.

```python

%%sql sqlite://
SELECT RCH, YR, MO, MAX(FLOW_INcms) as MAX_FLOWIN
FROM rch
GROUP BY RCH
HAVING MAX_FLOWIN > 3000.0
ORDER BY YR DESC, MO
```
